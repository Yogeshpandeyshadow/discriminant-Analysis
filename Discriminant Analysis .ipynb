```python
import pandas as pd
from sklearn.metrics import confusion_matrix
filepath="C:/Users/DELL/OneDrive/Desktop/IIT DHANBAD SAMRAT/7. Multivariate Lab/Practice/Exp 9 Dicriminant Analysis Data.xlsx"
df=pd.read_excel(filepath)
df
```

```python
df.drop(['Respond ent Number'], axis = 1, inplace=True)
```

```python
df.head()
```

```python

df.dropna(inplace=True)
```

```python
df.head()
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python

clf = LinearDiscriminantAnalysis()

X = df.iloc[:,1:].copy()
X.head()
```

```python

visit = df['Resort visit'].copy()
visit.head()
```

```python
result = clf.fit(X, visit)
print("\n Classes: ", result.classes_)
print("\n GROUP MEANS: \n", result.means_.round(4))
print("\n PRIOR PROBABILITY", result.priors_.round(4))
print("\n PREDICTIONS: ", result.predict(X))
print("\n CANONICAL DISCRIMINANT FUNCTION \n", result.scalings_.round(3))
```

```python
results=confusion_matrix(visit, result.predict(X))
```

```python
print("\nThe Confusion Matrix is: \n", results)
```

```python
score=result.score(X, visit)
```

```python
print("\n The accuracy is: ", score.round(4))
```
